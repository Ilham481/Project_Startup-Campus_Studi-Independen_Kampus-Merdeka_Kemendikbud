```python
from sklearn.preprocessing import LabelEncoder
label_attr = LabelEncoder()
data.Employee_Name  = label_attr.fit_transform(data.Employee_Name)

label_EmpID = LabelEncoder()
data.EmpID  = label_EmpID.fit_transform(data.EmpID)

data.head()
```

```python
#Ordinal Encoder
#untuk 'Gender', 'Education_Level', 'Marital_Status', 'Income_Category', 'Card_Category'
from sklearn.preprocessing import OrdinalEncoder

data['RecruitmentSource'].unique()
```

```python
enc_1 = OrdinalEncoder(categories=[['LinkedIn', 'Indeed', 'Google Search', 'Employee Referral',
       'Diversity Job Fair', 'On-line Web application', 'CareerBuilder',
       'Website', 'Other']], unknown_value=99, handle_unknown='use_encoded_value')

#ubah dulu sebagai array
y = np.array(data['RecruitmentSource']).reshape(-1,1)
enc_1.fit(y)

data['RecruitmentSource'] = enc_1.transform(y)
```
