```python
import torch
import torch.nn as nn
```

```python
# define a model in pytorch

# method-1

model = nn.Sequential(
    nn.Linear(28*28,256),
    nn.ReLU(),
    nn.Linear(256,64),
    nn.ReLU(),
    nn.Linear(64, 10)
)
```

```python
class MYModel(nn.Module) :
  def __init__(self) :
    super().__init__()
    self.l1 = nn.Linear(28*28, 256)
    self.l2 = nn.Linear(256,64)
    self.l3 = nn.Linear(64,10)
    self.do = nn.Dropout(0.1)
  
  def forward(self,x) :
    k1 = nn.functional.ReLU(self.l1(x))
    k2 = nn.functional.ReLU(self.l2(k1))
    do = self.do(k2+k1)
    y_hat =self.l3(do)
    return y_hat
    
modell1 = MYModel()
```

```python
# Loss Function

loss = nn.CrossEntropyLoss()
```

```python
# Optimiser

from torch import optim

params = model.parameters()
optimiser = optim.SGD(params,lr = 1e-2)
```

```python
from torchvision import datasets, transforms
from torch.utils.data import random_split, DataLoader

train_data = datasets.MNIST('data', train = True, download=True, transform = transforms.ToTensor())

train, val = random_split(train_data, [55000, 5000])

train_loader = DataLoader(train, batch_size = 32)
val_loader = DataLoader(val, batch_size = 32)
```

```python
#Training Loop

nb_epochs = 20

for epoch in range(nb_epochs) :
  losses = list()
  for batch in train_loader:
    x,y = batch
    # x ---> Batch-size x Channel x H x W
    b = x.size(0)
    x = x.view(b, -1)

    # forward
    y_hat = model(x)

    # compute the objective function
    J = loss(y_hat, y)

    #Cleaning the gradient
    model.zero_grad()

    # Accumulation partial derivative
    J.backward()

    optimiser.step()
    losses.append(J.item())
  print(f'Epoch: {epoch+1}, train loss: {torch.tensor(J).mean():.2f}')
```
